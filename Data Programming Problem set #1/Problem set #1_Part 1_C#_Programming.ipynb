<h4>Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. <br/>By Heron's formula, the area
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c), where
s = (a + b + c) /2.
</h4>

```
class triangleArea{
    
    public float calculateArea (float a, float b, float c){
        
        //To calculate area of triange all sides must be greater than 0 and sum of two sides must be greater than or equal third side
        if (a < 0 || b < 0 || c <0 || (a + b <= c) || a + c <=b || b + c <=a)
        {
            Console.Write("Invalid triangle");
        }
        
        double s = (a + b + c) / 2;
        return (float)Math.Sqrt(s * (s - a) * (s - b) * (s - c));
    }
}
```

```
triangleArea TA = new triangleArea();
Console.WriteLine(TA.calculateArea(2,2,2));
```

using System.Collections.Generic;

<h4>a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False depending on whether the point (x,y) lies in the rectangle with lower left corner (x1,y1) and upper right corner (x2,y2).</h4>

```
class checkPoints
{
    public bool inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        //To lie in the rectange value of x must be greater than x1 and less than x2 and 
            //similarly for y, must be greater than y1 and less than y2
        if (x > x1 && x < x2 && y > y1 && y < y2)
        {
            return true;
        }
        else
        {
            return false;
        }
    }
}
```

```
checkPoints CP = new checkPoints();
Console.WriteLine(CP.inside(1,1,0,0,2,3));
```

```
Console.WriteLine(CP.inside(-1,-1,0,0,2,3));
```
